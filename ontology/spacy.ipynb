```python
import spacy
import math
nlp=spacy.load("en_core_web_sm")


def extract_nouns(row):
  nouns = []
  print(row['sentence'])
  if isinstance(row['sentence'], str):
    doc=nlp(row['sentence'])
    for w in doc:
      if w.pos_ == "NOUN": #or x.pos_ == "PROPN" or x.pos_=="PRON"
        nouns.append(w)
  nouns

df['nouns'] = df.apply(extract_nouns, axis=1)
```
